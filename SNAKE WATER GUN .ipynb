```python
PLAYER1=0
PLAYER2=0
while True:
    
    print("WELCOME TO SNAKE WATER GUN GAME\n")
    print("press S for SANKE")
    print("press W for WATER")
    print("press G for GUN")

    x=input("enter your player1 choice:\n")
    y=input("enter your player2 choice:\n")
    if x==y:
        print("ITS DRAW\n")
        PLAYER1+=1
        PLAYER2+=1
        print("PLAYER1 SCORE:",PLAYER1)
        print("PLAYER2 SCORE:",PLAYER2)
        
    elif x=="S" and y=="W":
        if PLAYER2>=1:
            PLAYER2+=-1
            
            
        print("PLATER1 WINS")
        PLAYER1+=2
        
        print("PLAYER1 SCORE:",PLAYER1)
        print("PLAYER2 SCORE:",PLAYER2)
        
        
    elif x=="S" and y=="G":
        if PLAYER1>=1:
            PLAYER1+=-1
            
        print("PLATER2 WINS")
        
        PLAYER2+=2
        print("PLAYER1 SCORE:",PLAYER1)
        print("PLAYER2 SCORE:",PLAYER2)
        
    elif x=="W" and y=="S":
        if PLAYER1>=1:
            PLAYER1+=-1
        print("PLATER2 WINS")
        
        PLAYER2+=2
        print("PLAYER1 SCORE:",PLAYER1)
        print("PLAYER2 SCORE:",PLAYER2)
        
        
    elif x=="W" and y=="G":
        if PLAYER2>=1:
            PLAYER2+=-1
            
        print("PLATER1 WINS")
        PLAYER1+=2
     
        print("PLAYER1 SCORE:",PLAYER1)
        print("PLAYER2 SCORE:",PLAYER2)
        
    elif x=="G" and y=="S":
        if PLAYER2>=1:
            PLAYER2+=-1
        
        print("PLATER1 WINS")
        PLAYER1+=2
        
        print("PLAYER1 SCORE:",PLAYER1)
        print("PLAYER2 SCORE:",PLAYER2)
        
    elif x=="G" and y=="W":
        if PLAYER1>=1:
            PLAYER1+=-1
        print("PLATER2 WINS")
        
        PLAYER2+=2
        print("PLAYER1 SCORE:",PLAYER1)
        print("PLAYER2 SCORE:",PLAYER2)
        
    else:
        print("INVALID CHOICE\n")
    
        

        




```
